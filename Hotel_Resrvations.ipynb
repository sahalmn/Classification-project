```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Hotel Reservations.csv",nrows=10000)
df
```

```python
df["type_of_meal_plan"].value_counts()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.drop(["Booking_ID"],axis=1,inplace=True)
df.drop(["arrival_date"],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python

df["type_of_meal_plan"]=le.fit_transform(df["type_of_meal_plan"])
df["room_type_reserved"]=le.fit_transform(df["room_type_reserved"])
df["market_segment_type"]=le.fit_transform(df["market_segment_type"])
df["booking_status"]=le.fit_transform(df["booking_status"])
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE(random_state=1)
x_os,y_os=oversample.fit_resample(x,y)
```

```python
y_os.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x_os)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_os,test_size=0.2,random_state=1)
```

```python
# use--feature selection correlation
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
from sklearn.svm import SVC
sv=SVC()
from sklearn.metrics import classification_report
```

```python
lst=[knn,lr,nb,dt,sv]
for i in lst:
  print(i)
  i.fit(x_train,y_train)#knn.fit(x_train,y_train),lr.fit()
  y_pred=i.predict(x_test)
  print(classification_report(y_test,y_pred))


```

```python
#hyper parameter
from sklearn.model_selection import GridSearchCV
p={"criterion":["gini","entropy","log_loss"],
    "splitter":["best","random"],
    "max_depth":[2,3,4,5,6]} # Changed "max_depth" values to a list of integers
gr=GridSearchCV(dt,p,cv=10,scoring="accuracy") # Corrected the typo in "scoring"
gr.fit(x_train,y_train)
```

```python
print(gr.best_params_)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt1=DecisionTreeClassifier(criterion='gini', max_depth=6, splitter='best')
dt1.fit(x_train,y_train)
y_pred1=dt1.predict(x_test)
print(classification_report(y_test,y_pred1))
```

```python
import joblib
joblib.dump(dt1,"model.joblib")
joblib.dump(scaler,"scaler.joblib")
```

```python
%%writefile app.py
import streamlit as st
import joblib
import pandas as pd

# Streamlit app
st.set_page_config(page_title="Hotel Reservations", layout="centered")
st.title("🏨 Hotel Reservation Cancellation Predictor")
st.write("Please enter reservation details below:")

# Load the model and scaler
model = joblib.load("model.joblib")
scaler = joblib.load("scaler.joblib")

# Input fields
no_of_adults = st.number_input("No of Adults", min_value=0, max_value=10, step=1)
no_of_children = st.number_input("No of Children", min_value=0, max_value=5, step=1)
no_of_weekend_nights = st.number_input("No of Weekend Nights", min_value=0, max_value=10, step=1)
no_of_week_nights = st.number_input("No of Week Nights", min_value=0, max_value=100, step=1)
type_of_meal_plan = st.selectbox("Meal Plan", ["Meal Plan 1", "Meal Plan 2", "Meal Plan 3", "Not Selected"])
required_car_parking_space = st.selectbox("Car Parking Required?", ["No", "Yes"])
room_type_reserved = st.selectbox("Room Type Reserved", ["Room_Type 1", "Room_Type 2", "Room_Type 3", "Room_Type 4"])
lead_time = st.number_input("Lead Time (days)", min_value=0, max_value=365, step=1)
arrival_year = st.number_input("Arrival Year", min_value=2017, max_value=2018, step=1)
arrival_month = st.number_input("Arrival Month", min_value=1, max_value=12, step=1)
market_segment_type = st.selectbox("Market Segment Type", ["Online", "Offline", "Corporate", "Aviation", "Complementary"])
repeated_guest = st.number_input("Repeated Guest (0 = No, 1 = Yes)", min_value=0, max_value=1, step=1)
no_of_previous_cancellations = st.number_input("No of Previous Cancellations", min_value=0, max_value=20, step=1)
no_of_previous_bookings_not_canceled = st.number_input("No of Previous Bookings Not Canceled", min_value=0, max_value=50, step=1)
avg_price_per_room = st.number_input("Avg Price Per Room", min_value=0.0, max_value=10000.0, step=1.0)
no_of_special_requests = st.number_input("No of Special Requests", min_value=0, max_value=10, step=1)

# Categorical mapping
meal_plan_map = {
    "Not Selected": 0,
    "Meal Plan 1": 1,
    "Meal Plan 2": 2,
    "Meal Plan 3": 3
}
car_parking_map = {"No": 0, "Yes": 1}
room_type_map = {
    "Room_Type 1": 0,
    "Room_Type 2": 1,
    "Room_Type 3": 2,
    "Room_Type 4": 3
}
market_segment_map = {
    "Online": 0,
    "Offline": 1,
    "Corporate": 2,
    "Aviation": 3,
    "Complementary": 4
}

# Apply mapping
meal_plan_value = meal_plan_map[type_of_meal_plan]
car_parking_value = car_parking_map[required_car_parking_space]
room_type_value = room_type_map[room_type_reserved]
market_segment_value = market_segment_map[market_segment_type]

# Create input data
input_data = pd.DataFrame([{
    'lead_time': lead_time,
    'arrival_year': arrival_year,
    'arrival_month': arrival_month,
    'no_of_adults': no_of_adults,
    'no_of_children': no_of_children,
    'no_of_weekend_nights': no_of_weekend_nights,
    'no_of_week_nights': no_of_week_nights,
    'type_of_meal_plan': meal_plan_value,
    'required_car_parking_space': car_parking_value,
    'room_type_reserved': room_type_value,
    'market_segment_type': market_segment_value,
    'repeated_guest': repeated_guest,
    'no_of_previous_cancellations': no_of_previous_cancellations,
    'no_of_previous_bookings_not_canceled': no_of_previous_bookings_not_canceled,
    'avg_price_per_room': avg_price_per_room,
    'no_of_special_requests': no_of_special_requests
}])

# Scale if required
try:
    input_scaled = scaler.transform(input_data)
except:
    input_scaled = input_data  # Skip scaling if scaler is not necessary

# Prediction
if st.button("Predict Cancellation"):
    try:
        prediction = model.predict(input_scaled)
        if prediction[0] == 1:
            st.error("⚠️ This reservation is likely to be **canceled**.")
        else:
            st.success("✅ This reservation is likely to be **honored**.")
    except Exception as e:
        st.error(f"An error occurred during prediction: {e}")
```
